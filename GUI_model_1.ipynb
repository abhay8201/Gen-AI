```python
import tkinter as tk
from tkinter import scrolledtext
from langchain_groq import ChatGroq

# Initialize the ChatGroq model
llm = ChatGroq(
    model="mixtral-8x7b-32768",
    temperature=0,
    groq_api_key="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"
)

# Function to get the response from the model
def getresponse(a):
    response = llm.invoke(a)
    return response.content

# Function to send the message and display it in the chat window
def send_message():
    user_input = user_entry.get().strip()  # Get the user input from the entry box
    if user_input == "":
        return  # Prevent sending empty messages
    
    chat_window.config(state=tk.NORMAL)  # Enable editing the chat window
    
    # Display user input in blue
    chat_window.insert(tk.END, "You: " + user_input + "\n", "user")

    # Get the model response
    model_response = getresponse(user_input)
    
    # Display bot response in green
    chat_window.insert(tk.END, "Bot: " + model_response + "\n\n", "bot")
    
    # Scroll the chat window to the bottom
    chat_window.yview(tk.END)
    
    # Clear the input field for the next question
    user_entry.delete(0, tk.END)
    
    chat_window.config(state=tk.DISABLED)  # Disable editing after updating chat

# Create the main window
root = tk.Tk()
root.title("Chat with Groq Model")
root.configure(bg="#222831")  # Dark background

# Create a scrolled text widget for displaying the chat
chat_window = scrolledtext.ScrolledText(root, width=60, height=20, wrap=tk.WORD, state=tk.DISABLED, bg="#EEEEEE", font=("Arial", 12))
chat_window.grid(row=0, column=0, padx=10, pady=10)

# Add text tags for styling
chat_window.tag_configure("user", foreground="blue", font=("Arial", 12, "bold"))
chat_window.tag_configure("bot", foreground="green", font=("Arial", 12, "bold"))

# Create an entry widget for user input
user_entry = tk.Entry(root, width=50, font=("Arial", 12))
user_entry.grid(row=1, column=0, padx=10, pady=10)

# Create a button to send the message
send_button = tk.Button(root, text="Send", width=15, command=send_message, font=("Arial", 12, "bold"), bg="#00ADB5", fg="white")
send_button.grid(row=2, column=0, padx=10, pady=10)

# Run the application
root.mainloop()
```
